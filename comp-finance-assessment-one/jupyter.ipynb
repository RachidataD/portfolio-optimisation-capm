## Advanced Options

Beyond local editing two things are quite important: editing the notebooks in the markdown format and running Jupyter remotely.
The latter matters when we want to run the code on a faster server.
The former matters since Jupyter's native ipynb format stores a lot of auxiliary data that is
irrelevant to the content,
mostly related to how and where the code is run.
This is confusing for Git, making
reviewing contributions very difficult.
Fortunately there is an alternative---native editing in the markdown format.

### Markdown Files in Jupyter

If you wish to contribute to the content of this book, you need to modify the
source file (md file, not ipynb file) on GitHub.
Using the notedown plugin we
can modify notebooks in the md format directly in Jupyter.


First, install the notedown plugin, run the Jupyter Notebook, and load the plugin:

```
pip install d2l-notedown  # You may need to uninstall the original notedown.
jupyter notebook --NotebookApp.contents_manager_class='notedown.NotedownContentsManager'
```

You may also turn on the notedown plugin by default whenever you run the Jupyter Notebook.
First, generate a Jupyter Notebook configuration file (if it has already been generated, you can skip this step).

```
jupyter notebook --generate-config
```

Then, add the following line to the end of the Jupyter Notebook configuration file (for Linux or macOS, usually in the path `~/.jupyter/jupyter_notebook_config.py`):

```
c.NotebookApp.contents_manager_class = 'notedown.NotedownContentsManager'
```

After that, you only need to run the `jupyter notebook` command to turn on the notedown plugin by default.

### Running Jupyter Notebooks on a Remote Server

Sometimes, you may want to run Jupyter notebooks on a remote server and access it through a browser on your local computer. If Linux or macOS is installed on your local machine (Windows can also support this function through third-party software such as PuTTY), you can use port forwarding:

```
ssh myserver -L 8888:localhost:8888
```

The above string `myserver` is the address of the remote server.
Then we can use http://localhost:8888 to access the remote server `myserver` that runs Jupyter notebooks. We will detail on how to run Jupyter notebooks on AWS instances
later in this appendix.

```
pip install jupyter_contrib_nbextensions
```

````python
import pandas as pd
import numpy as np
from tabulate import tabulate # Library for advanced formatting
# installing appropriate libraries for according code functions

def simulate_wealth_growth(initial_deposit: float, interest_rate: float, periods: int, additional_rate: float):
# professional description of function - docstring

    """
    Simulates the increase of a clients wealth over time and monitors total balance, earned interest and fixed additional deposits over a fixed number of periods.

    This function shows financial simulation logic by applying compound interest and a fixed periodic addition.
    Args:
        initial_deposit (float): The clientâ€™s starting amount.
        interest_rate (float): The interest rate per period (as a decimal, e.g., 0.05).
        periods (int): The total number of periods to track.
        additional_rate (float): Additional deposit per period, expressed as a
                                 percentage of the initial_deposit.

    Returns:
        pd.DataFrame: A table containing the Period, Total Balance,
                      Additional Deposit, and Interest Earned for each period.
    """

    # robustness, the input Validity
    # robust financial code needs to verify that inputs make financial sense.
    if any(val < 0 for val in [initial_deposit, interest_rate, periods, additional_rate]):
        print("Error: All financial inputs must be non-negative. Simulation stopped.")
        return

    # individualisation, an advanced error check
    # this block prevents user error where they enter '5' instead of '0.05'.
    if interest_rate > 1:
        print("Warning: Interest rate is > 1. Converting percentage to decimal for simulation accuracy.")
        interest_rate = interest_rate / 100

    # formula for the fixed absolute amount of additional deposit
    fixed_additional_deposit = initial_deposit * additional_rate

    # initialising variables for the simulation
    current_balance = initial_deposit
    period_list = range(0, periods + 1)
    #intialise the results lists with period 0, the starting point
    # custom variable names
    total_balance = [current_balance] # monitors total value overtime
    periodic_interest = [0.0] # monitors periodic interest earned
    periodic_cash_inflow = [0.0] # monitors deposit added each period

    # core simulation loop,the engine
    # looping from period 1 to last period defined in the 'periods' parameter
    # ... (the rest of your function logic follows)
    for t in period_list:
        if t == 0:
            continue # skip calculation for period 0 (initial state)

        # calculate interest earned
        # interest is calculated on the previous period's balance.
        interest = total_balance[t-1] * interest_rate

        # calculate the new balance

        # crucial step where the interest and the fixed deposit are added.
        # logic: balance_t = balance_{t-1} + interest_t + fixed_Deposit
        new_balance = total_balance[t-1] + interest + fixed_additional_deposit

        # store the results for the current period
        periodic_interest.append(interest)
        periodic_cash_inflow.append(fixed_additional_deposit)
        total_balance.append(new_balance)


    # customising the output (the presentation)

    # creating the DataFrame using the custom lists (
    df = pd.DataFrame({
        'Period': list(period_list),
        'Total Balance': total_balance,
        'Additional Deposit': periodic_cash_inflow,
        'Interest Earned': periodic_interest
    })

    # format the monetary values for clean output to two decimal places
    df_formatted = df.round(2)

    # prepare data for the custom 'tabulate' table style
    table_data = df_formatted.values.tolist()
    headers = df_formatted.columns.tolist()

    # custom table style
    print("\nWealth Growth Simulation (Custom Output Style):")
    print(tabulate(table_data, headers=headers, tablefmt="simple", floatfmt=",.2f"))

    # the function returns the DataFrame
    return df
#execution and testing block as it didnt run previously (must be placed outside the function)

# calls the function using the exact required inputs from the assignment brief
# (initial_deposit=1000, interest_rate=0.05, periods=5, additional_rate=0.1)

print("\n--- Testing against Assignment Brief Example ---")
simulate_wealth_growth(
    initial_deposit=1000,
    interest_rate=0.05,
    periods=5,
    additional_rate=0.1
)

#the secondary test case
# add a new test case to demonstrate the robustness and custom style of the code.
print("\n--- Running Secondary Test Case (Demonstrating Robustness) ---")
simulate_wealth_growth(
    initial_deposit=5000,
    interest_rate=0.02,
    periods=10,
    additional_rate=0.05
)
````

````python
# installing necessary library to download real stock and index price data
!pip install yfinance --quiet

# importing new libraries for financial data work
import yfinance as yf
import pandas as pd
import numpy as np

# this block uses the yfinance library to connect to a financial API, to download historical market data for analysis.

def fetch_financial_data(ticker_stock: str, ticker_index: str, start_date: str, end_date: str) -> pd.DataFrame:
    """
    Retrieves daily adjusted (or close) prices for a chosen stock and a market index
    using the yfinance API and returns a clean 2-column DataFrame.
    """
    tickers = [ticker_stock, ticker_index]
    try:
        # Keep Adj Close available (old behavior) and reduce chatter
        data = yf.download(
            tickers, start=start_date, end=end_date,
            auto_adjust=False, progress=False
        )

        if data.empty:
            print(f"No data returned for {tickers}. Check tickers/date range.")
            return pd.DataFrame()

        # yfinance returns a MultiIndex when multiple tickers are used.
        # Prefer 'Adj Close' if present, otherwise fall back to 'Close'.
        if isinstance(data.columns, pd.MultiIndex):
            lvl0 = data.columns.get_level_values(0)
            if 'Adj Close' in lvl0:
                px = data['Adj Close']
            elif 'Close' in lvl0:
                px = data['Close']
            else:
                print(f"'Adj Close'/'Close' not found in columns: {data.columns}")
                return pd.DataFrame()
        else:
            # Single-index columns (unlikely with 2 tickers, but keep for safety)
            if 'Adj Close' in data.columns:
                px = data[['Adj Close']]
            elif 'Close' in data.columns:
                px = data[['Close']]
            else:
                print(f"'Adj Close'/'Close' not found in columns: {data.columns}")
                return pd.DataFrame()

        # Reorder to match our [stock, index] list and rename for clarity
        px = px[tickers]  # ensures column order is [ticker_stock, ticker_index]
        px.columns = [f'{ticker_stock}_Price', f'{ticker_index}_Price']

        # Clean
        return px.dropna(how='any')

    except Exception as e:
        print(f"Error retrieving data for {tickers}: {e}")
        return pd.DataFrame()

#choosing assets
# to diversify, choose an asset with traditionally low market correlation
INDEX_TICKER = '^NDX' # Nasdaq 100 is another common market index

# at least two years of daily data required
# note, ensure the START_DATE is two years prior to the END_DATE
START_DATE = '2022-01-01' # Changed start date
END_DATE = '2024-01-01' # Changed end date to a date in the past

# the data retrieval function
price_data = fetch_financial_data('MSFT', '^NDX', START_DATE, END_DATE)

# verify data exists before proceeding
if price_data.empty:
    raise ValueError("No price data downloaded. Please check tickers or date range.")

# print the data head for successful download
print(f"\nSuccessfully downloaded {len(price_data)} days of data.")
print("--- Price Data Head ---")
print(price_data.head())

def compute_financial_metrics(df: pd.DataFrame) -> dict:
    """
    Computes all required financial metrics (returns, volatility, correlation, etc.)
    from the daily adjusted closing prices.

    Args:
        df (pd.DataFrame): DataFrame containing daily price data for the stock and index.

    Returns:
        dict: A dictionary containing all calculated metrics for easy reporting.
    """

    # daily Returns
    # daily Returns are the first step in financial analysis as they normalise the data.
    # daily returns, the required metric
    # daily returns are calculated as the percentage change in the adjusted close price.
    returns_df = df.pct_change().dropna()

    # separating the returns for convenience and display formatting
    stock_returns = returns_df.iloc[:, 0] # First column is the stock returns
    index_returns = returns_df.iloc[:, 1] # Second column is the index returns

    # summary statistics, a required metric
    # generating summary statistics as min/max demonstrate range and mean and standard deviation demonstrate central tendency and risk.
    summary_stats = returns_df.describe().T

    # annualised statistics, a required metric.
    ANNUALIZATION_FACTOR = 252  # annualisation factor (standard convention)

    # annualised mean Return: mean daily return * annualisation factor
    annual_return_stock = stock_returns.mean() * ANNUALIZATION_FACTOR
    annual_return_index = index_returns.mean() * ANNUALIZATION_FACTOR

    # annualised volatility (risk measure): std dev * sqrt(annualisation factor)
    annual_vol_stock = stock_returns.std() * np.sqrt(ANNUALIZATION_FACTOR)
    annual_vol_index = index_returns.std() * np.sqrt(ANNUALIZATION_FACTOR)

    # correlation, a required metric
    correlation = stock_returns.corr(index_returns)

    # beta calculation
    beta = correlation * (annual_vol_stock / annual_vol_index)

    # display all final results into a dictionary for practicality in the outline of the final report
    metrics = {
        'Stock_Ticker': df.columns[0].replace('_Price', ''),
        'Index_Ticker': df.columns[1].replace('_Price', ''),
        'Annualized_Return_Stock': annual_return_stock,
        'Annualized_Vol_Stock': annual_vol_stock,
        'Annualized_Return_Index': annual_return_index,
        'Annualized_Vol_Index': annual_vol_index,
        'Correlation_Stock_Index': correlation,
        'Beta': beta,
        'Daily_Returns_DataFrame': returns_df,
        'Summary_Statistics_Returns': summary_stats
    }

    return metrics

#the metrics function
financial_results = compute_financial_metrics(price_data)

# print the final numerical results for the report, f-strings for clean percentage output
print("\n--- Final Financial Metrics (For Report Appendix) ---")
print(pd.Series({
    'Asset Ticker': financial_results['Stock_Ticker'],
    'Index Ticker': financial_results['Index_Ticker'],
    'Stock Annual Return': f"{financial_results['Annualized_Return_Stock']:.2%}",
    'Index Annual Return': f"{financial_results['Annualized_Return_Index']:.2%}",
    'Stock Annual Volatility': f"{financial_results['Annualized_Vol_Stock']:.2%}",
    'Index Annual Volatility': f"{financial_results['Annualized_Vol_Index']:.2%}",
    'Correlation (Stock vs Index)': f"{financial_results['Correlation_Stock_Index']:.4f}",
    'Beta (Sensitivity to Market)': f"{financial_results['Beta']:.4f}"
}))

print("\n--- Summary Statistics of Daily Returns (Required Table) ---")
# display the summary statistics table
print(financial_results['Summary_Statistics_Returns'].round(5))
````

````python
import matplotlib.pyplot as plt
#matplotlib is the standard library used to create all the required professional charts for the report.

def generate_visualizations(df_prices: pd.DataFrame, df_returns: pd.DataFrame, stock_ticker: str, index_ticker: str):
    """
    Generates required charts: Price Plot (with Moving Average), Returns Histogram, and Scatterplot.

    Args:
        df_prices (pd.DataFrame): DataFrame of adjusted closing prices.
        df_returns (pd.DataFrame): DataFrame of daily returns.
        stock_ticker (str): The symbol for the individual asset.
        index_ticker (str): The symbol for the market index.
    """

    # calcuating moving average for the line chart plot
    # using a 50 day window since it was explicitly asked for moving averages.
    window_size = 50
    ma_stock = df_prices[f'{stock_ticker}_Price'].rolling(window=window_size).mean()
    ma_index = df_prices[f'{index_ticker}_Price'].rolling(window=window_size).mean()

    #chart A: trends, plotting prices with moving averages.
    # --- Chart A: Fixed Price Plot with Dual Y-Axes (Trends) ---
    plt.figure(figsize=(15, 6))

    # creating the first Y-AXIS (for the Index, the high-value series)
    ax1 = plt.gca() # Get the current axis
    ax1.plot(df_prices[f'{index_ticker}_Price'], label=f'{index_ticker} Price', color='salmon', linewidth=2)
    ax1.set_xlabel('Date')
    ax1.set_ylabel(f'{index_ticker} Adjusted Close Price', color='salmon')
    ax1.tick_params(axis='y', labelcolor='salmon')

    # Create the SECOND Y-AXIS (for the Stock, the lower-value series)
    ax2 = ax1.twinx()

    # Plotting the stock price and the 50-day moving average (MA) on the new axis
    ax2.plot(df_prices[f'{stock_ticker}_Price'], label=f'{stock_ticker} Price', color='gray', alpha=0.8, linewidth=1)
    ax2.plot(ma_stock, label=f'{stock_ticker} 50-Day MA', color='blue', linewidth=2)
    ax2.set_ylabel(f'{stock_ticker} Adjusted Close Price', color='blue')
    ax2.tick_params(axis='y', labelcolor='blue')

    # Final chart settings
    plt.title(f'Dual Axis Price Time Series: {stock_ticker} vs. {index_ticker}')
    ax1.grid(True, linestyle='--', alpha=0.6)

    # Combine legends from both axes
    lines, labels = ax1.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax1.legend(lines + lines2, labels + labels2, loc='upper left')

    # FLAG: INDIVIDUALIZATION: ANNOTATION (Add your event here)
    # Example: ax2.annotate('Tech Rally', xy=('2023-01-01', 250), xytext=('2022-10-01', 270), arrowprops=dict(facecolor='black', shrink=0.05))

    plt.show()

    # simultaneusly plotting the price and the 50-day moving average
    plt.plot(df_prices[f'{stock_ticker}_Price'], label=f'{stock_ticker} Price', color='gray', alpha=0.5, linewidth=1)
    plt.plot(ma_stock, label=f'{stock_ticker} 50-Day MA', color='blue', linewidth=2)
    plt.plot(df_prices[f'{index_ticker}_Price'], label=f'{index_ticker} Price', color='red', alpha=0.5, linewidth=1)

    plt.title(f'Price Time Series and 50-Day Moving Average: {stock_ticker} vs. {index_ticker}')
    plt.xlabel('Date')
    plt.ylabel('Adjusted Close Price')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)

    # annotation, important aspect of visualisation.
    # Example: plt.annotate('Major Price Peak', xy=('2024-03-01', 200), xytext=('2024-01-01', 220), arrowprops=dict(facecolor='black', shrink=0.05))
    plt.show()

    # chart b, a returns histogram for volatility and risk
    plt.figure(figsize=(10, 6))
    plt.hist(df_returns.iloc[:, 0], bins=50, alpha=0.6, label=stock_ticker, color='darkgreen', density=True)
    plt.hist(df_returns.iloc[:, 1], bins=50, alpha=0.6, label=index_ticker, color='red', density=True)
    plt.title('Histogram of Daily Returns (Risk Comparison)')
    plt.xlabel('Daily Return')
    plt.ylabel('Frequency (Probability Density)')
    plt.legend()
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.show()

    # chart c: Scatterplot for correlation and or sensitivity
    plt.figure(figsize=(8, 6))
    plt.scatter(df_returns.iloc[:, 1], df_returns.iloc[:, 0], alpha=0.5, color='purple')
    plt.title(f'Scatterplot: {stock_ticker} Returns vs. {index_ticker} Returns (Correlation Visual)')
    plt.xlabel(f'{index_ticker} Daily Returns')
    plt.ylabel(f'{stock_ticker} Daily Returns')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()
````
