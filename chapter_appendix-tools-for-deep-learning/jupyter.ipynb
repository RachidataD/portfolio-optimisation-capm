## Advanced Options

Beyond local editing two things are quite important: editing the notebooks in the markdown format and running Jupyter remotely.
The latter matters when we want to run the code on a faster server.
The former matters since Jupyter's native ipynb format stores a lot of auxiliary data that is
irrelevant to the content,
mostly related to how and where the code is run.
This is confusing for Git, making
reviewing contributions very difficult.
Fortunately there is an alternative---native editing in the markdown format.

### Markdown Files in Jupyter

If you wish to contribute to the content of this book, you need to modify the
source file (md file, not ipynb file) on GitHub.
Using the notedown plugin we
can modify notebooks in the md format directly in Jupyter.


First, install the notedown plugin, run the Jupyter Notebook, and load the plugin:

```
pip install d2l-notedown  # You may need to uninstall the original notedown.
jupyter notebook --NotebookApp.contents_manager_class='notedown.NotedownContentsManager'
```

You may also turn on the notedown plugin by default whenever you run the Jupyter Notebook.
First, generate a Jupyter Notebook configuration file (if it has already been generated, you can skip this step).

```
jupyter notebook --generate-config
```

Then, add the following line to the end of the Jupyter Notebook configuration file (for Linux or macOS, usually in the path `~/.jupyter/jupyter_notebook_config.py`):

```
c.NotebookApp.contents_manager_class = 'notedown.NotedownContentsManager'
```

After that, you only need to run the `jupyter notebook` command to turn on the notedown plugin by default.

### Running Jupyter Notebooks on a Remote Server

Sometimes, you may want to run Jupyter notebooks on a remote server and access it through a browser on your local computer. If Linux or macOS is installed on your local machine (Windows can also support this function through third-party software such as PuTTY), you can use port forwarding:

```
ssh myserver -L 8888:localhost:8888
```

The above string `myserver` is the address of the remote server.
Then we can use http://localhost:8888 to access the remote server `myserver` that runs Jupyter notebooks. We will detail on how to run Jupyter notebooks on AWS instances
later in this appendix.

### Timing

We can use the `ExecuteTime` plugin to time the execution of each code cell in Jupyter notebooks.
Use the following commands to install the plugin:

```
pip install jupyter_contrib_nbextensions
jupyter contrib nbextension install --user
jupyter nbextension enable execute_time/ExecuteTime
```

````python
import pandas as pd
import numpy as np
from tabulate import tabulate # Library for advanced formatting
# installing appropriate libraries for according code functions

def simulate_wealth_growth(initial_deposit: float, interest_rate: float, periods: int, additional_rate: float):
# professional description of function - docstring

    """
    Simulates the increase of a clients wealth over time and monitors total balance, earned interest and fixed additional deposits over a fixed number of periods.

    This function shows financial simulation logic by applying compound interest and a fixed periodic addition.
    Args:
        initial_deposit (float): The clientâ€™s starting amount.
        interest_rate (float): The interest rate per period (as a decimal, e.g., 0.05).
        periods (int): The total number of periods to track.
        additional_rate (float): Additional deposit per period, expressed as a
                                 percentage of the initial_deposit.

    Returns:
        pd.DataFrame: A table containing the Period, Total Balance,
                      Additional Deposit, and Interest Earned for each period.
    """

    # robustness, the input Validity
    # robust financial code needs to verify that inputs make financial sense.
    if any(val < 0 for val in [initial_deposit, interest_rate, periods, additional_rate]):
        print("Error: All financial inputs must be non-negative. Simulation stopped.")
        return

    # individualisation, an advanced error check
    # this block prevents user error where they enter '5' instead of '0.05'.
    if interest_rate > 1:
        print("Warning: Interest rate is > 1. Converting percentage to decimal for simulation accuracy.")
        interest_rate = interest_rate / 100

    # formula for the fixed absolute amount of additional deposit
    fixed_additional_deposit = initial_deposit * additional_rate

    # initialising variables for the simulation
    current_balance = initial_deposit
    period_list = range(0, periods + 1)
    #intialise the results lists with period 0, the starting point
    # custom variable names
    total_balance = [current_balance] # monitors total value overtime
    periodic_interest = [0.0] # monitors periodic interest earned
    periodic_cash_inflow = [0.0] # monitors deposit added each period

    # core simulation loop,the engine
    # looping from period 1 to last period defined in the 'periods' parameter
    # ... (the rest of your function logic follows)
    for t in period_list:
        if t == 0:
            continue # skip calculation for period 0 (initial state)

        # calculate interest earned
        # interest is calculated on the previous period's balance.
        interest = total_balance[t-1] * interest_rate

        # calculate the new balance

        # crucial step where the interest and the fixed deposit are added.
        # logic: balance_t = balance_{t-1} + interest_t + fixed_Deposit
        new_balance = total_balance[t-1] + interest + fixed_additional_deposit

        # store the results for the current period
        periodic_interest.append(interest)
        periodic_cash_inflow.append(fixed_additional_deposit)
        total_balance.append(new_balance)


    # customising the output (the presentation)

    # creating the DataFrame using the custom lists (
    df = pd.DataFrame({
        'Period': list(period_list),
        'Total Balance': total_balance,
        'Additional Deposit': periodic_cash_inflow,
        'Interest Earned': periodic_interest
    })

    # format the monetary values for clean output to two decimal places
    df_formatted = df.round(2)

    # prepare data for the custom 'tabulate' table style
    table_data = df_formatted.values.tolist()
    headers = df_formatted.columns.tolist()

    # custom table style
    print("\nWealth Growth Simulation (Custom Output Style):")
    print(tabulate(table_data, headers=headers, tablefmt="simple", floatfmt=",.2f"))

    # the function returns the DataFrame
    return df
# --- Execution and Testing Block (MUST be placed OUTSIDE the function) ---

# FLAG: CUSTOM EXAMPLE CALL
# This calls your function using the exact required inputs from the assignment brief
# (initial_deposit=1000, interest_rate=0.05, periods=5, additional_rate=0.1)

print("\n--- Testing against Assignment Brief Example ---")
simulate_wealth_growth(
    initial_deposit=1000,
    interest_rate=0.05,
    periods=5,
    additional_rate=0.1
)

# FLAG: INDIVIDUALIZATION - SECONDARY TEST CASE
# Add a new test case to demonstrate the robustness and custom style of your code.
print("\n--- Running Secondary Test Case (Demonstrating Robustness) ---")
simulate_wealth_growth(
    initial_deposit=5000,
    interest_rate=0.02,
    periods=10,
    additional_rate=0.05
)
````
